```python
lesioanados_2020 = lesioanados_2020.drop(columns= ['type'])

lesioanados_2020
```

```python
lesioanados_2020['fechas_lesion'] = pd.to_datetime(lesioanados_2020['fechas_lesion'])

lesioanados_2020
```

```python
lesionados_2020 = lesioanados_2020.sort_values(by=['id', 'fechas_lesion'], ascending=[False, False])

lesionados_2020
```

```python
lesionados_2020.reset_index(drop=True, inplace=True)

lesionados_2020
```

```python
lesionados_2020 = lesionados_2020[~lesionados_2020['reason'].isin(['Rest', 'Suspended'])]

lesionados_2020
```

```python
lesionados_2020.reset_index(drop=True, inplace=True)

lesionados_2020
```

```python
lesionados_2020['fechas_lesion'][0] 
```

```python
time_diff_per_id = lesionados_2020.groupby('id')['fechas_lesion'].agg(lambda x: x.max() - x.min())

time_diff_per_id
```

```python
lesionados_2020 = pd.merge(lesionados_2020, time_diff_per_id, on='id', how='left')

lesionados_2020
```

```python
lesionados_2020 = lesionados_2020.rename(columns={'fechas_lesion_x': 'fechas_lesion', 'fechas_lesion_y': 'tiempo_lesion'})

lesionados_2020
```

```python
lesionados_2020.reset_index(drop=True, inplace=True)

lesionados_2020
```
