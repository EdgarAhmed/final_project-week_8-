## Crear DF de las plantillas de los años:

2020-21

2021-22

2022-23

```python
import pymongo
import pandas as pd
from bson import ObjectId
```

```python
client = pymongo.MongoClient("mongodb://localhost:27017")
db = client['plantillas_barsa_2010']
collection = db['plantillas_atm_2020']
```

## Plantilla 2020

```python
#plantilla 2020

target_id = ObjectId("656df8c7cff3b4031c5118d4") #Cambiando el Id de cada uno, tendré la plantilla correspondiente. 
query = {"_id": target_id}
data_from_mongo20 = collection.find_one(query)
```

```python
data_from_mongo20
```

```python
data_from_mongo2020 = [data_from_mongo20] if data_from_mongo20 else []

data_from_mongo2020
```

```python
for e in range(len(data_from_mongo2020)):
    for i in range(len(data_from_mongo2020[e])):
        print(data_from_mongo2020[e]['response'][i]['player'])
        print(data_from_mongo2020[e]['response'][i]['statistics'])
```

```python
data_list = []

for e in range(len(data_from_mongo2020)):
    for i in range(len(data_from_mongo2020[e]['response'])):
        for a in range (len(data_from_mongo2020[e]['response'][i]['statistics'])):
            player_data = data_from_mongo2020[e]['response'][i]['player']
            statistics_data = data_from_mongo2020[e]['response'][i]['statistics'][a]
            other = data_from_mongo2020[e]['response'][i]['statistics'][a]['team']

            # Extract relevant information
            player_info = {
                'id': player_data['id'],
                'name': player_data['name'],
                'firstname': player_data['firstname'],
                'age': player_data['age'],
                'birth_date': player_data['birth']['date'],
                'nationality': player_data['nationality'],
                'height': player_data['height'],
                'weight': player_data['weight'],
                'injured': player_data['injured'],
                'league': statistics_data['league']['name'],
                'team': statistics_data['team']['name'],
                'season': statistics_data['league']['season'],
                'games_appearances': statistics_data['games']['appearences'],
                'games_lineups': statistics_data['games']['lineups'],
                'games_minutes': statistics_data['games']['minutes'],
                'games_position': statistics_data['games']['position'],
                #'champios': statistics_data[0]['team']['league']['name']
            }

            data_list.append(player_info)

# Convert the list of dictionaries to a Pandas DataFrame
df = pd.DataFrame(data_list)

# Display the DataFrame
plantilla_2020 = df

plantilla_2020
```
