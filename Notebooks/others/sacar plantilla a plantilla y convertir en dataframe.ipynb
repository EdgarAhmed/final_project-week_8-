```python
import pymongo
import pandas as pd
from bson import ObjectId
```

```python
# Replace the placeholders with your MongoDB connection details
client = pymongo.MongoClient("mongodb://localhost:27017")
db = client['plantillas_barsa_2010']
collection = db['plantillas_barsa_2010']
```

```python
# Replace 'your_collection_name' with the name of your MongoDB collection
#collection = db['plantillas_barsa_2010']

# Retrieve all documents from the collection
#data_from_mongo = list(collection.find_one(sort=[("_id", 1)]))

#data_from_mongo
```

```python
target_id = ObjectId("65688386c42e904eaef804af") #Cambiando el Id de cada uno, tendré la plantilla correspondiente. 
query = {"_id": target_id}
data_from_mongo = collection.find_one(query)

```

```python
type(data_from_mongo)
```

```python
data_from_mongo_2 = [data_from_mongo] if data_from_mongo else []

data_from_mongo_2
```

```python
data_from_mongo_2[0]['response'][0]['statistics']
```

```python
for e in range(len(data_from_mongo_2)):
    for i in range(len(data_from_mongo_2[e])):
        print(data_from_mongo_2[e]['response'][i]['player'])
        print(data_from_mongo_2[e]['response'][i]['statistics'])
```

```python
data_list = []

for e in range(len(data_from_mongo_2)):
    for i in range(len(data_from_mongo_2[e]['response'])):
        player_data = data_from_mongo_2[e]['response'][i]['player']
        statistics_data = data_from_mongo_2[e]['response'][i]['statistics']

        # Extract relevant information
        player_info = {
            'id': player_data['id'],
            'name': player_data['name'],
            'firstname': player_data['firstname'],
            'age': player_data['age'],
            'birth_date': player_data['birth']['date'],
            'nationality': player_data['nationality'],
            'height': player_data['height'],
            'weight': player_data['weight'],
            'injured': player_data['injured'],
            'league': statistics_data[0]['league']['name'],
            'team': statistics_data[0]['team']['name'],
            'season': statistics_data[0]['league']['season'],
            'games_appearances': statistics_data[0]['games']['appearences'],
            'games_lineups': statistics_data[0]['games']['lineups'],
            'games_minutes': statistics_data[0]['games']['minutes'],
            'games_position': statistics_data[0]['games']['position'],
            #'champios': statistics_data[0]['team']['league']['name']
        }

        data_list.append(player_info)

# Convert the list of dictionaries to a Pandas DataFrame
df = pd.DataFrame(data_list)

# Display the DataFrame
plantilla_2011 = df

plantilla_2011
```
