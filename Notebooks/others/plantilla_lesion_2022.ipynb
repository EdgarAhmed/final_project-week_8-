```python
import pandas as pd
from datetime import datetime
import numpy as np


pd.set_option('display.max_rows', 10)
```

```python
plantilla_bar_2022['league'].unique()
```

```python
plantilla_bar_2022['league'] = plantilla_bar_2022['league'].replace({'Friendlies Clubs': 'Club Friendlies'})

plantilla_bar_2022
```

```python
plantilla_bar_2022['league'].unique()
```

```python
plantilla_bar_2022.drop(columns=['team'])
```

```python
#Fecha de nacimiento con el año que corresponde 2022

plantilla_bar_2022['birth_date'] = pd.to_datetime(plantilla_bar_2022['birth_date'])

plantilla_bar_2022['birth_date'] = plantilla_bar_2022['birth_date'].fillna(datetime.now())

plantilla_bar_2022
```

```python
current_date = datetime(2021, 12, 31)

age_timedelta = current_date - plantilla_bar_2022['birth_date']
plantilla_bar_2022['age'] = (age_timedelta / pd.Timedelta(days=365.25)).astype(int)


plantilla_bar_2022
```

```python
#Suma total de los minutos por partido

grouped_df = plantilla_bar_2022.groupby(['id', 'league'])['games_minutes'].sum().reset_index()

# Pivot the DataFrame to have 'id' as index, 'league' as columns, and 'games_minutes' as values
pivoted_df = grouped_df.pivot(index='id', columns='league', values='games_minutes').reset_index()

# Fill NaN values with 0
pivoted_df = pivoted_df.fillna(0)

# Create a new column with the total sum across the four leagues
pivoted_df['total_minutes'] = pivoted_df[['Club Friendlies', 'La Liga', 'UEFA Champions League', 'Copa del Rey', 'UEFA Europa League']].sum(axis=1)

# Convert 'total_minutes' to int64
pivoted_df['total_minutes'] = pivoted_df['total_minutes'].astype('int64')

# Merge the new DataFrame with the original DataFrame based on 'id'
final_df = pd.merge(plantilla_bar_2022, pivoted_df[['id', 'total_minutes']], on='id', how='left')

# Display the final DataFrame
plantilla_bar_2022 = final_df

plantilla_bar_2022 
```

```python
unique_terms = ['Injury', 'Illness', 'Muscle Injury', 'Foot Injury',
       'Thigh Injury', 'Ankle Injury', 'Shoulder Injury', 'Rest',
       'Knee Injury', 'Calf Injury', 'Hamstring Injury']


lesionados_2022['unique_reason_count'] = lesionados_2022.groupby('id')['reason'].transform(lambda x: x[x.isin(unique_terms)].nunique())

lesionados_2022
```

```python
lesionados_2022 = lesionados_2022.rename(columns={'unique_reason_count': 'numero_lesiones_jugador'})

lesionados_2022.reset_index(drop=True, inplace=True)

lesionados_2022
```

```python

def add_times_appeared(lesionados_2022):
    # List of terms in the column 'reason'
    terms = ['Injury', 'Illness', 'Muscle Injury', 'Foot Injury',
       'Thigh Injury', 'Ankle Injury', 'Shoulder Injury', 'Rest',
       'Knee Injury', 'Calf Injury', 'Hamstring Injury']

    # Create a new column 'count_appeared' and initialize it to 0
    lesionados_2022['count_appeared'] = 0

    # Iterate over each term and update the 'count_appeared' column
    for term in terms:
        # Filter rows where 'reason' is equal to the current term
        term_rows = lesionados_2022[lesionados_2022['reason'] == term]

        # Update the count based on the number of unique 'fechas_lesion' for the term
        lesionados_2022.loc[lesionados_2022['reason'] == term, 'count_appeared'] = term_rows['fechas_lesion'].nunique()

    # Add 2 to the 'count_appeared' column where there is only one 'fechas_lesion'
    lesionados_2022.loc[lesionados_2022['fechas_lesion'].notna() & (lesionados_2022['count_appeared'] == 1), 'count_appeared'] += 2

    return lesionados_2022

# Example usage:
# Assuming 'your_df' is your original DataFrame
lesionados_2022 = add_times_appeared(lesionados_2022)
```
