```python

def get_lesiones_info():
    # Connect to the MongoDB server
    client = MongoClient('mongodb://localhost:27017/')

    # Access the database and collection
    db = client['plantillas_barsa_2010']
    collection = db['lesiones_barsa_20']

    # Define the projection to include only specific fields
    projection = {
        'id': 1,
        'name': 1,
        'photo': 1,
        'reason': 1,
        'date': 1,
        'season': 1
    }

    # Execute the query with the specified projection
    result = collection.find({}, projection)

    # Create a list to store the results
    lesions_info = []

    # Append each document to the list
    for document in result:
        lesions_info.append(document)

    # Close the MongoDB connection
    client.close()

    return lesions_info

# Call the function and print the results
lesiones_data = get_lesiones_info()
for lesion in lesiones_data:
    print(lesion)
```
