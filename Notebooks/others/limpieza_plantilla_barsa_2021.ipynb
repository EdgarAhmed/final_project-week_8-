```python
import pandas as pd
from datetime import datetime
import numpy as np


pd.set_option('display.max_rows', 10)
```

```python
plantilla_bar_2021 = pd.read_csv ('/Users/edgarahmed/Desktop/ironhack/semana_8/modelo predictivo/plantillas_lesionados_barsa_csv/plantilla_barsa_2021.csv')

plantilla_bar_2021
```

```python
plantilla_bar_2021['league'].unique()
```

```python
plantilla_bar_2021['league'] = plantilla_bar_2021['league'].replace({'Friendlies Clubs': 'Club Friendlies'})

plantilla_bar_2021
```

```python
plantilla_bar_2021.drop(columns=['team'])
```

```python
plantilla_bar_2021['birth_date'] = pd.to_datetime(plantilla_bar_2021['birth_date'])

plantilla_bar_2021
```

```python
plantilla_bar_2021['birth_date'] = plantilla_bar_2021['birth_date'].fillna(datetime.now())

plantilla_bar_2021
```

```python
current_date = datetime(2020, 12, 31)

age_timedelta = current_date - plantilla_bar_2021['birth_date']
plantilla_bar_2021['age'] = (age_timedelta / pd.Timedelta(days=365.25)).astype(int)


plantilla_bar_2021
```

```python
grouped_df = plantilla_bar_2021.groupby(['id', 'league'])['games_minutes'].sum().reset_index()

# Pivot the DataFrame to have 'id' as index, 'league' as columns, and 'games_minutes' as values
pivoted_df = grouped_df.pivot(index='id', columns='league', values='games_minutes').reset_index()

# Fill NaN values with 0
pivoted_df = pivoted_df.fillna(0)

# Create a new column with the total sum across the four leagues
pivoted_df['total_minutes'] = pivoted_df[['La Liga', 'Copa del Rey', 'UEFA Champions League', 'Club Friendlies']].sum(axis=1)

# Convert 'total_minutes' to int64
pivoted_df['total_minutes'] = pivoted_df['total_minutes'].astype('int64')

# Merge the new DataFrame with the original DataFrame based on 'id'
final_df = pd.merge(plantilla_bar_2021, pivoted_df[['id', 'total_minutes']], on='id', how='left')

# Display the final DataFrame
plantilla_bar_2021 = final_df

plantilla_bar_2021 
```

```python
unique_terms = ['Calf Injury', 'Lacking Match Fitness', 'Knee Injury', 'Muscle Injury', "Coach's decision",
                'Tendon Injury', 'Red Card', 'Back Injury', 'Stomach Disorder', 'Thigh Injury', 'Heart Problems',
                'Leg Injury', 'Injury', 'Illness', 'Yellow Cards', 'Hand Injury', 'Inactive', 'Ankle Injury', 'Groin Injury']


lesionados_2021['unique_reason_count'] = lesionados_2021.groupby('id')['reason'].transform(lambda x: x[x.isin(unique_terms)].nunique())

lesionados_2021
```

```python
filtered_df = lesionados_2021[lesionados_2021['reason'].isin(unique_terms)]

# Count unique occurrences for each id
lesiones_totales_barsa = filtered_df.groupby('reason')['id'].nunique()

lesiones_totales_barsa
```

```python
total_unique_occurrences = filtered_df.groupby('reason')['id'].nunique().sum()

total_unique_occurrences
```
