```python
plantilla_bar_2020['birth_date'] = pd.to_datetime(plantilla_bar_2020['birth_date'])

plantilla_bar_2020
```

```python
plantilla_bar_2020['birth_date'] = plantilla_bar_2020['birth_date'].fillna(datetime.now())

plantilla_bar_2020
```

```python
current_date = datetime(2020, 12, 31)

age_timedelta = current_date - plantilla_bar_2020['birth_date']
plantilla_bar_2020['age'] = (age_timedelta / pd.Timedelta(days=365.25)).astype(int)


plantilla_bar_2020
```

```python
import pandas as pd

# Assuming plantilla_bar_2020 is your DataFrame
# Replace 'column_name' with the actual name of your 'id', 'league', and 'games_minutes' columns

# Group by 'id' and 'league' and calculate the sum of 'games_minutes'
grouped_df = plantilla_bar_2020.groupby(['id', 'league'])['games_minutes'].sum().reset_index()

# Pivot the DataFrame to have 'id' as index, 'league' as columns, and 'games_minutes' as values
pivoted_df = grouped_df.pivot(index='id', columns='league', values='games_minutes').reset_index()

# Fill NaN values with 0
pivoted_df = pivoted_df.fillna(0)

# Create a new column with the total sum across the four leagues
pivoted_df['total_minutes'] = pivoted_df[['La Liga', 'Copa del Rey', 'UEFA Champions League', 'Club Friendlies']].sum(axis=1)

# Convert 'total_minutes' to int64
pivoted_df['total_minutes'] = pivoted_df['total_minutes'].astype('int64')

# Merge the new DataFrame with the original DataFrame based on 'id'
final_df = pd.merge(plantilla_bar_2020, pivoted_df[['id', 'total_minutes']], on='id', how='left')

# Display the final DataFrame
plantilla_bar_2020 = final_df

plantilla_bar_2020 
```
