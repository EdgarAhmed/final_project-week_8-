## Lesionados Barsa (de la API)

```python
import pymongo
import pandas as pd
from bson import ObjectId

client = pymongo.MongoClient("mongodb://localhost:27017")
db = client['plantillas_barsa_2010']
collection = db['lesiones_barsa_20']
```

## Lesionados temp 2020-21

```python
target_id = ObjectId("656886b8c42e904eaef804b9") #Cambiando el Id de cada uno, tendr√© la plantilla correspondiente. 
query = {"_id": target_id}
data_from_mongo = collection.find_one(query)
```

```python
type(data_from_mongo)
```

```python
data_from_mongo_2 = [data_from_mongo] if data_from_mongo else []


data_from_mongo_2
```

```python
for e in data_from_mongo_2[0]['response']:

    print(e['player'])
```

```python
data = []

for e in range(len(data_from_mongo_2)):
    for i in range(len(data_from_mongo_2[e]['response'])):
        
            
        player_data = data_from_mongo_2 [e]['response'][i]['player']
        team_data = data_from_mongo_2 [e] ['response'] [i] ['team']
        fixture  = data_from_mongo_2 [e] ['response'] [i] ['fixture']
        league = data_from_mongo_2 [e] ['response'] [i] ['league']

        player_info = {
            'id':player_data ['id'],
            'name': player_data['name'],
            'type': player_data ['type'],
            'reason': player_data ['reason'],
            'team': team_data ['name'],
            'fixture': fixture ['date'],
            'league': league ['name']
        }

        data.append(player_info)

# Create a DataFrame
df = pd.DataFrame(data)

# Print the DataFrame
lesionados_2020 = df

lesionados_2020
```
